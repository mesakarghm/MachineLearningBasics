```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
plt.rcParams['figure.figsize'] = (20.0,10.0)

#reading the data
data = pd.read_csv('headbrain.csv')

#printing data shape and data head 
print(data.shape)
data.head()
```

```python
#collecting x and y 
#consists of headsize and y consists of brain weight
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

#total number of values 
m = len(X)

#using formula to calculate value of b1(m) and b0(c) 
numer = 0 
denom = 0
for i in range(m): 
    numer +=(X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) **2
b1 = numer / denom 
b0 = mean_y - (b1 * mean_x)

#print coefficients
print(b1, b0)
```

```python
#Plotting values and Regression line 
max_x = np.max(X) + 100 
min_x = np.min(X) - 100 

#calculating line values x and y 
x = np.linspace(min_x, max_x, 1000)
y = b0 + b1 * x

#plotting line 
plt.plot(x,y, color = '#58b977', label = 'Regression Line')

#plotting scatter plots 
plt.scatter(X,Y, color = '#ef5423', label = 'Scatter Plot')

plt.xlabel('Head Size in cm3')
plt.ylabel("Brain weight in grams")
plt.legend()
plt.show()
```

```python
#calculating model fit using R2 values 

ss_t = 0 
ss_r = 0 
for i in range(m): 
    y_pred = b0 + b1 * X[i]
    ss_t += (Y[i] - mean_y) ** 2
    ss_r += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_r / ss_t)
print(r2)
```
